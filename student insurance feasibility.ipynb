### Estimating Claims
To estimate claims for the Delhi student insurance pilot, the expected number of claims is calculated by applying an assumed claim rate to the total number of enrolled students. The claim rate represents the proportion of insured students who are likely to file at least one claim during the year. This gives the expected number of claimants. 

If the analysis also considers multiple claims by the same student, the expected claimants are further multiplied by claim frequency to estimate the total number of claims submitted. This projected claim volume is then used to calculate expected claims cost by multiplying it with the average claim amount derived from the historical dataset.

### Formula

Expected Claims = Students × Claim Rate

If we also want to account for repeated claims by the same claimant:

Projected Total Claims = Number of Students × Claim Rate × Claim Frequency


Where:

- **Number of Students** = total students enrolled in the pilot
- **Claim Rate** = percentage of insured students expected to file at least one claim
- **Claim Frequency** = average number of claims filed by each claimant


### Example Calculation

Assuming:

- Students = 2,000
- Claim Rate = 10%
- Claim Frequency = 1.25


Expected *Claimants* = 2000 × 10% = 200

Projected Total Claims = 200 ×1.25 = 250

```python
import matplotlib.pyplot as plt
import numpy as np

# Apply clean visual style
plt.style.use("seaborn-v0_8-whitegrid")

# Data
years = ["Year 1", "Year 2", "Year 3"]
claims = np.array([18597826.00, 22317391.20, 26780869.44]) / 1e6
profit = np.array([-6697826.00, -8037391.20, -9644869.44]) / 1e6
revenue = np.array([14000000.00, 16800000.00, 20160000.00]) / 1e6

# Premium Color Palette
COLOR_REV = "#1f77b4"  # Royal Blue
COLOR_CLAIMS = "#d62728"  # Crimson
COLOR_PROFIT = "#d9381e"  # Dark Red/Coral for Negative Profit

fig, ax = plt.subplots(figsize=(10, 6.5), dpi=150)

# 1. Plot Lines
ax.plot(
    years,
    revenue,
    marker="o",
    markersize=8,
    linewidth=2.5,
    color=COLOR_REV,
    label="Revenue",
)

ax.plot(
    years,
    claims,
    marker="s",
    markersize=8,
    linewidth=2.5,
    color=COLOR_CLAIMS,
    label="Claims Cost",
)

ax.plot(
    years,
    profit,
    marker="^",
    markersize=8,
    linewidth=2.5,
    linestyle="--",
    color=COLOR_PROFIT,
    label="Profit (Net Deficit)",
)

# 2. Highlight Loss Area (Underwriting Gap)
ax.fill_between(
    years,
    revenue,
    claims,
    color=COLOR_CLAIMS,
    alpha=0.10,
    label="Underwriting Loss Gap",
)

# 3. Reference Lines & Grid
ax.axhline(0, color="#333333", linewidth=1.2, linestyle="-", zorder=1)
ax.grid(True, linestyle=":", alpha=0.6, color="#cccccc")

# 4. Data Point Annotations with Dynamic Placement
for x_idx, (y_rev, y_clm, y_prf) in enumerate(
    zip(revenue, claims, profit)
):
    # Revenue Label
    ax.text(
        x_idx,
        y_rev - 1.2,
        f"₹{y_rev:.2f}M",
        color=COLOR_REV,
        ha="center",
        va="top",
        fontsize=9.5,
        fontweight="bold",
    )

    # Claims Label
    ax.text(
        x_idx,
        y_clm + 1.2,
        f"₹{y_clm:.2f}M",
        color=COLOR_CLAIMS,
        ha="center",
        va="bottom",
        fontsize=9.5,
        fontweight="bold",
    )

    # Profit Label
    ax.text(
        x_idx,
        y_prf - 1.2,
        f"₹{y_prf:.2f}M",
        color=COLOR_PROFIT,
        ha="center",
        va="top",
        fontsize=9.5,
        fontweight="bold",
    )

# 5. Titles & Labels
ax.set_title(
    "3-Year Projected Financial Performance",
    fontsize=16,
    fontweight="bold",
    pad=15,
    color="#222222",
)
ax.set_xlabel("Projection Timeline", fontsize=11, fontweight="bold", labelpad=10)
ax.set_ylabel("Amount (₹ Millions)", fontsize=11, fontweight="bold", labelpad=10)

# Set comfortable y-axis limits to prevent text clipping
ax.set_ylim(-14, 32)

# Spines cleanup
for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)

# Legend Configuration
ax.legend(
    loc="upper left",
    frameon=True,
    facecolor="white",
    edgecolor="#e0e0e0",
    fontsize=10,
)

plt.tight_layout()
plt.show()
```

### 4. Loss Ratio

**Formula**

| $$\text{Loss Ratio} = \frac{\text{Claims Cost}}{\text{Premium Revenue}} \times 100$$ |
----------------------------------------------------------------------------------------

*Loss Ratio is a key financial metric used by insurance companies to measure how much of the premium revenue is spent on paying claims.*

| Interpretation |
------------------
| - *Loss Ratio < 100%* : Premium revenue is greater than claims paid. The insurance scheme is generating enough income to cover claims.                                                                                                                            |
| - *Loss Ratio = 100%* : Every rupee collected as premium is used to pay claims, leaving no margin for administrative expenses or profit.                                                                                                                            |
| - *Loss Ratio > 100%* : Claims exceed premium revenue, meaning the insurer loses money even before considering operational costs.    |

Year-wise Loss Ratio

| Year | Total Revenue (₹) | Claims Cost (₹) | Loss Ratio (%) |
| ---- | ----------------: | --------------: | -------------: |
| 1    |     14,000,000.00 |   18,597,826.00 |    **132.84%** |
| 2    |     16,800,000.00 |   22,317,391.20 |    **132.84%** |
| 3    |     20,160,000.00 |   26,780,869.44 |    **132.84%** |

The projected loss ratio of 132.84% across all three years indicates that the insurance scheme is not financially sustainable under the current premium, claim frequency, and average claim amount.

```python
import matplotlib.pyplot as plt
import numpy as np

# Apply clean visual style
plt.style.use("seaborn-v0_8-whitegrid")

# Data setup
years = ["Year 1", "Year 2", "Year 3"]
loss_ratio = [132.84] * len(years)

# Figure setup
fig, ax = plt.subplots(figsize=(9, 5.5), dpi=150)

# 1. Fill Loss Region Above 100% Threshold
ax.fill_between(
    years,
    100,
    loss_ratio,
    color="#ff9900",
    alpha=0.18,
    label="Loss Zone (>100%)",
)

# 2. Main Trend Line
ax.plot(
    years,
    loss_ratio,
    marker="o",
    linewidth=2.5,
    markersize=8,
    color="#e65100",  # Rich Coral Orange
    label="Projected Loss Ratio",
    zorder=3,
)

# 3. 100% Threshold Line
ax.axhline(
    100,
    color="#d32f2f",  # Crimson Red
    linestyle="--",
    linewidth=1.8,
    label="100% Breakeven Threshold",
    zorder=2,
)

# 4. Data Labels
for x, y in zip(years, loss_ratio):
    ax.text(
        x,
        y + 2.5,
        f"{y:.2f}%",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold",
        color="#e65100",
    )

# 5. Axis Formatting & Spacing
ax.set_title(
    "Projected Loss Ratio over 3 Years",
    fontsize=15,
    fontweight="bold",
    pad=15,
    color="#222222",
)
ax.set_xlabel("Projection Period", fontsize=11, fontweight="bold", labelpad=10)
ax.set_ylabel("Loss Ratio (%)", fontsize=11, fontweight="bold", labelpad=10)

# Set generous bounds so labels aren't squeezed
ax.set_ylim(85, 150)

# Grid setup
ax.grid(True, linestyle=":", alpha=0.6, color="#cccccc")

# Hide chart border spines for a modern look
for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)

# Legend Configuration
ax.legend(
    loc="lower right",
    frameon=True,
    facecolor="white",
    edgecolor="#e0e0e0",
    fontsize=9.5,
)

plt.tight_layout()
plt.show()
```

## Conclusion

The three-year financial projection indicates that the proposed student health insurance scheme is **not financially sustainable under the current assumptions**. Although premium revenue is projected to increase each year due to a 20% annual growth in student enrollment, the expected claims cost increases proportionally and consistently exceeds the revenue generated.

The projected **loss ratio of 132.84%** across all three years indicates that for every **₹100 collected in premiums, approximately ₹132.84 is paid out in claims**, excluding administrative expenses. Consequently, the scheme records a negative profit each year, with losses increasing from **₹6.70 million in Year 1** to **₹9.64 million in Year 3**. The return on investment (ROI) also remains negative throughout the projection period.

These findings suggest that the current premium structure is insufficient to cover the expected healthcare expenditure. To improve the financial viability of the scheme, adjustments such as increasing premium rates, reducing claim costs through cost-control measures, lowering claim frequency via preventive healthcare initiatives, or adopting a combination of these strategies should be considered. Such measures would help achieve a sustainable balance between premium income and claims expenditure while ensuring long-term stability of the insurance program.
