# Estimating Claims

To estimate claims for the **Delhi Student Insurance Pilot**, the expected number of claims is calculated by applying an assumed claim rate to the total number of enrolled students. The **Claim Rate** represents the proportion of insured students likely to file at least one claim during the year, yielding the **Expected Claimants**.

When accounting for multiple claims filed by the same student, expected claimants are multiplied by **Claim Frequency** to estimate total claim volume. This total volume is then multiplied by the average claim amount (derived from historical dataset EDA) to calculate projected total claims cost.


### Formulas

> **Expected Claimants:**
> 
> $$\text{Expected Claimants} = \text{Number of Students} \times \text{Claim Rate}$$
> 
> 

> **Projected Total Claims (accounting for repeat claims):**
> 
> $$\text{Projected Total Claims} = \text{Number of Students} \times \text{Claim Rate} \times \text{Claim Frequency}$$
> 
> 

**Where:**

* **Number of Students:** Total students enrolled in the pilot program
* **Claim Rate:** Percentage of insured students expected to file $\ge 1$ claim
* **Claim Frequency:** Average number of claims submitted per claimant


### Example Calculation

**Baseline Assumptions:**

* **Students ($N$):** $2,000$
* **Claim Rate ($R$):** $10\%$
* **Claim Frequency ($F$):** $1.25$

#### 1. Calculate Expected Claimants

$$\text{Expected Claimants} = 2,000 \times 10\% = \mathbf{200 \text{ claimants}}$$

#### 2. Calculate Projected Total Claims

$$\text{Projected Total Claims} = 200 \times 1.25 = \mathbf{250 \text{ total claims}}$$

### 4. Loss Ratio Analysis

>**Formula**
> $$\text{Loss Ratio} = \left( \frac{\text{Claims Cost}}{\text{Premium Revenue}} \right) \times 100$$

*Loss Ratio is a key financial metric used by insurance companies to measure the proportion of premium revenue consumed by claim payouts.*

**Interpretation Guide** 

* **Loss Ratio < 100%:** Premium revenue exceeds total claims paid. The insurance scheme generates sufficient income to cover claims and operational overhead.
* **Loss Ratio = 100%:** Every rupee collected in premiums is spent paying claims, leaving zero margin for administrative costs or profits.
* **Loss Ratio > 100%:** Claims exceed total premium revenue, resulting in an underwriting loss before factoring in operational expenses.

**Year-wise Loss Ratio Projections**

| Year | Total Revenue (₹) | Claims Cost (₹) | Loss Ratio (%) |
| :--- | ---: | ---: | ---: |
| **Year 1** | ₹14,000,000.00 | ₹18,597,826.00 | **132.84%** |
| **Year 2** | ₹16,800,000.00 | ₹22,317,391.20 | **132.84%** |
| **Year 3** | ₹20,160,000.00 | ₹26,780,869.44 | **132.84%** |


> ⚠️ **Key Takeaway:** The projected loss ratio of **132.84%** across all three years indicates that the insurance scheme is financially unsustainable under current premium rates, claim frequencies, and average claim amounts.

## Conclusion

The three-year financial projection indicates that the proposed student health insurance scheme is **not financially sustainable under the current assumptions**. Although premium revenue is projected to increase each year due to a 20% annual growth in student enrollment, the expected claims cost increases proportionally and consistently exceeds the revenue generated.

The projected **loss ratio of 132.84%** across all three years indicates that for every **₹100 collected in premiums, approximately ₹132.84 is paid out in claims**, excluding administrative expenses. Consequently, the scheme records a negative profit each year, with losses increasing from **₹6.70 million in Year 1** to **₹9.64 million in Year 3**. The return on investment (ROI) also remains negative throughout the projection period.

These findings suggest that the current premium structure is insufficient to cover the expected healthcare expenditure. To improve the financial viability of the scheme, adjustments such as increasing premium rates, reducing claim costs through cost-control measures, lowering claim frequency via preventive healthcare initiatives, or adopting a combination of these strategies should be considered. Such measures would help achieve a sustainable balance between premium income and claims expenditure while ensuring long-term stability of the insurance program.
